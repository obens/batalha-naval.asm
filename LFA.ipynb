```python
def pre_process(lines):
    for idx in range(len(lines)):
        lines[idx] = re.sub('\n', '', lines[idx])
    return lines

c_term = 1

#funcao para adicionar mais simbolos na lista do alfabeto
def countTerm():
  counter = 0
  global c_term
  if alfa_count >= 24 * c_term  :
    for l in string.ascii_lowercase:
      alfab.append(l * (c_term   + 1))
      char = l*(c_term + 1)
      lfaDict['simb'].append(char.upper())
      counter +=1
    c_term  += 1
  #completa o dicionario
  for key in lfaDict.keys():
    if key != 'simb':
      for i in range(counter):
        lfaDict[key].append('')

#remover simbolos especificos
def remove_symb(string, symbols):
  string = "".join(c for c in string if c not in symbols)
  return string

#separar o NT na gramatica
def separe(string):
  string = remove_symb(string, [' '])
  m = re.search(r"\<(\w+)\>", string)
  simb = m.group(1)
  var= re.sub(m.group(0), '', string)

  
  return var, simb

def findEnd(lines):
  for idx, line in enumerate(lines):
    if line == '':
      return idx
```

```python
#percorro todas as keys e vejo quais foram usadas pra criar um novo dicionario somente com oq ue foi usado

c_used = []
for c in columns:
  col = df[c]
  for line in col:
    if str(line) != '':
      #print(c)
      c_used.append(c)
      break

simbs = np.array(df.simb)
counter = 0 #apenas para ter uma profundidade
simbols = []
for s in simbs:
  if "*" in s:
    counter+=1
    simbols.append(s)
    continue
  arr = np.array(df[df.simb == s])
  #print(arr)
  for a in arr[0][1::]:
    if a != '':
      #print(s)
      counter+=1
      simbols.append(s)
      break
      
afnd = {}
afnd['simb'] = []
for c in c_used:
  afnd[c] = []

for key in afnd.keys():
  for i in range(counter):
    afnd[key].append('')
```

```python
def treat_(afnd, count, at_estado, simbols, c_used, df):

  estados_splt = at_estado.split(',')
  final =  False

  for estado in estados_splt:
    if '*'+estado in simbols:
      estado = '*'+estado
      if not final:
        final = True
        at_estado = '*'+at_estado

    initial = df[df.simb == estado]   
    #print(estado)
    #print(initial)

    afnd['simb'][count] = at_estado
    for c in c_used:
      #print(c)
      term = np.array(initial[c])[0]
      term = term.split(',')[1::]
      if len(term) > 1:
        #state = ",".join(s for s in term)
        afnd[c][count] = state + afnd[c][count] 
        #count+=1
        continue

      for t in term:
        afnd[c][count] = t + ','+afnd[c][count] 

    dict_ = pd.DataFrame.from_dict(afnd)
    initial = dict_[dict_.simb == at_estado]

  for c in c_used:
    term = np.array(initial[c])[0]
    if term == '':
      continue 
    #print('-----', term)
    if term[-1] == ',':
      term = term[0:-1]
    if term not in estados:
      #print("--------", term)
      estados.append(term)
```

```python
estados = []
estados.append('S')
count =  0

for estado in estados:

  if "," in estado:
    #print(estado)
    treat_(afnd, count, estado, simbols, c_used, df)

  else:
    if estado in simbols:
      idx_s = simbols.index(estado)
      if "*" in simbols[idx_s]:
        estado = '*'+estado

    elif '*'+estado in simbols:
      idx_s = simbols.index('*'+estado)
      if "*" in simbols[idx_s]:
        estado = '*'+estado


    initial = df[df.simb == estado]   
    #print(estado)
    #print(initial)

    afnd['simb'][count] = estado
    for c in c_used:
      #print(c)
      #print(initial)
      #print(estado)
      term = np.array(initial[c])[0]
      term = term.split(',')[1::]
      if len(term) > 1:
        state = ",".join(s for s in term)
        
        afnd[c][count] = state + afnd[c][count] 

        estados.append(state)
        #count+=1
        continue
      for t in term:
        afnd[c][count] = t + afnd[c][count] 
        if t not in estados:
          estados.append(t)
  count+=1
```

```python

ult_idx = 0
for idx, cel in enumerate(afnd['simb']):
  if cel == '':
    ult_idx = idx
    break

err = 'ERR'

if idx == len(afnd['simb'])-1:
  afnd['simb'].append('*'+err) 
  for key in afnd.keys():
    if key == 'simb':
      continue 
    afnd[key].append('')

else:
  for key in afnd.keys():
    for cel in range(len(afnd[key])):
      afnd[key] = afnd[key][0:ult_idx+1]

  afnd['simb'][ult_idx] = '*'+err 

for key in afnd.keys():
  if key == 'simb':
    continue
  for cel in range(len(afnd[key])):
    if afnd[key][cel] == '':
      afnd[key][cel] = err

```

```python
#Dict da GLC para retornar a Key (ESTADO) que da certa produção pertence e pre processa a lista de produções
glcDict = {}
prods = []
for i, line in enumerate(glc):
  line = line.split('->')
  line[1] = line[1][1::]
  prods.append(line[1])
  if line[0] not in glcDict.keys():
    glcDict[line[0]] = []
    glcDict[line[0]].append(line[1])
  else:
    glcDict[line[0]].append(line[1])
# retorna o ESTADO que certa produção pertence
def getProd(p):
  for k in glcDict.keys():
    for prod in glcDict[k]:
      if prod == p:
        return re.sub(' ', '', k)
```
